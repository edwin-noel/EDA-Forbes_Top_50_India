# <center> List of Largest Companies in India - 2022 || Forbes Global 2000 || EDA 

### Objective - Exploratory data analysis to find insights on Top 50 Largest Companies in India.

```python
#importing necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#ignoring warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing the dataset

df = pd.read_csv('2022_Forbes_list.csv')
```

```python
#checking the shape of the dataset

df.shape
```

```python
#checking the data

df.head()
```

```python
#checking if the Rank column is having unique values

df['Rank in India'].is_unique
```

```python
#information about the dataset

df.info()
```

We can see that there are no null values present in this dataset. 

Also, we can see a column named 'Unnamed: 0' which is nothing but the serial number, so we drop it. 

```python
#dropping the index column 

df.drop(columns= 'Unnamed: 0', axis = 1, inplace = True)
```

```python
#chekcing the above tast is performed successfully or not

df.info()
```

The columns 'Changes Rank from last Year' and 'Forbes 2000 rank in World' are of object datatype, this is because the columns have parentheses, so we will remove it and convert the datatype to int. 

```python
df['Changes Rank from last Year'] = df['Changes Rank from last Year'].str.replace('(','').str.replace(')','').astype('int64')

df['Changes Rank from last Year.1'] = df['Changes Rank from last Year.1'].str.replace('(','').str.replace(')','').astype('int64')
```

```python
#checking the task is successfully performed

df.head()
```

```python
df.info()
```

```python
#checking the unique values in the dataset

df.nunique()
```

```python
#dividing categorical columns and continious columns based on the datatype

cat_cols = [col for col in df.columns if df[col].dtype == 'object']

cont_cols = [col for col in df.columns if df[col].dtypes in ['int64', 'float64']]
```

```python
print("Categorical columns: ", cat_cols)

print("Continuous columns: ", cont_cols)
```

```python
#plotting the graph to see the Industries that made to the list of Largest companies in India List

plt.figure(figsize= (10,6))
sns.countplot(x= 'Industry', data = df)
plt.title('Industries in Top 50 Largest Companies in India List', fontsize = 20)
plt.xlabel('Industry Name', fontsize = 16)
plt.ylabel('Count', fontsize = 16)
plt.xticks(rotation = 90)
plt.grid(axis = 'y')
plt.show()
```

###### Observation- 

Most companies in the list are from the banking industry followed by the Oil and Gas and Infotech Industry.

```python
#checking headquaters of these Companies

plt.figure(figsize= (10,6))
sns.countplot(x = 'Headquarters', data = df)
plt.title('Headquarters of Top 50 Largest Companies in India', fontsize = 20)
plt.xlabel('Name of Places', fontsize = 16)
plt.ylabel('Count of Headquarters', fontsize = 16)
plt.xticks(rotation = 90)
plt.grid(axis = 'y')
plt.show() 
```

###### Observation - 

Most companies headquarters are in Mumbai followed by New Delhi.

```python
#checking headquaters of these Companies

plt.figure(figsize= (16, 8))
sns.countplot(x = 'Industry', hue ='Headquarters', data = df)
plt.title('Headquarters based on the Industry', fontsize = 20)
plt.xlabel('Name of Industry', fontsize = 16)
plt.ylabel('Count of Industry', fontsize = 16)
plt.xticks(rotation = 90)
plt.legend(loc='upper right')
plt.grid(axis = 'y')
plt.show() 
```

###### Observation - 

From the above plot, we can see that Mumbai have the highest number of Banking Headquarters, New Delhi have the highest numbers of Oil and Gas companies headquarters and Bangalore have the highest number of Infotech Companies Headquarters.

```python
#plotting a graph to see the Revenue, Profit, Assets and Value of the companies

cols = ['Revenue(billions US$)', 'Profit(billions US$)', 'Assets(billions US$)', 'Value(billions US$)']

def data_plt(cols_list):
    for col in cols:
        plt.figure(figsize = (16,6))
        plt.title('')
        sns.barplot(data = df, x = 'Name', y= col)
        plt.xlabel('Name of the Company', fontsize = 12)
        plt.ylabel(col, fontsize = 12)
        plt.xticks(rotation = 90)
        plt.grid(axis = 'y')
        plt.show()
        
data_plt(cols)
```

###### Observation

- Reliance Industries, Indian Oil Corporation and Oil and Natural Gas Corporation are the top three companies with the highest Revenue.

- Reliance Industry, Oil and Natural Gas Corporation and HDFC Bank have the highest Profit.

- State Bank of India, HDFC Bank and ICICI Bank have the highest assets. 

- Reliance Industries, Tata Consultancy Services and HDFC Bank have the most value. 

It seems that Banking, Oil & Gas and Infotech Industry is dominating the in the list of Largest Industry in India. 

```python
df.head()
```

```python
#adding up the column to check the previous year rank in India and in Forbes 2000 World Rank


df['Rank last year'] = df['Rank in India'] + df['Changes Rank from last Year']
df['Forbes 2000 Rank Last Year'] = df['Forbes 2000 rank in World'] + df['Changes Rank from last Year.1']

df.head()
```

```python
change_rank_india = df[['Name', 'Rank in India', 'Rank last year']].set_index('Name')

change_rank_india.head()
```

```python
change_rank_india.describe()
```

```python
#boxplot

plt.figure(figsize=(10,6))
plt.boxplot(change_rank_india)
plt.grid(axis = 'y')
plt.show()
```

We can see from the average mean that there is not much change in Company's Rank in India from last year's rank.

```python
#scatter plot to check the change in Company's Rank in India

plt.figure(figsize=(16,6))
sns.scatterplot(data = change_rank_india)
plt.xticks(rotation = 90)
plt.grid(axis = 'y')
plt.show()
```

```python
forbes_rank_change = df[['Name', 'Forbes 2000 rank in World', 'Forbes 2000 Rank Last Year']].set_index('Name')

forbes_rank_change.head()
```

```python
forbes_rank_change.describe()
```

```python
#boxplot 

plt.figure(figsize=(10,6))
plt.boxplot(forbes_rank_change)
plt.grid(axis = 'y')
plt.show()
```

We can see from the average (mean) that there was an improvement in the average Forbes Rank between the two years for companies in India.

```python
#scatter plot to check the change in Forbes Rank in two years

plt.figure(figsize=(16,6))
sns.scatterplot(data = forbes_rank_change)
plt.xticks(rotation = 90)
plt.grid(axis = 'y')
plt.show()
```

##### Final Observation:

From the above analysis, we can conclude that 

1. Banking Industry have the highest number in the list with the maximum headquarters in Mumbai.


2. Banking, Oil and Gas and Infotech Industries have the highest share in Revenue, Profit, Asset and Value. 


3. There is an improvement in the average Forbes 2000 Rank of the companies as compared to their last year's rank.
